```python
# pdf_extraction_compare.py

from pypdf import PdfReader
import pdfplumber
# import fitz  # PyMuPDF
# from unstructured.partition.pdf import partition_pdf
from pdfminer.high_level import extract_text


PDF_PATH = "Document Loaders/Documents/ML_research_paper.pdf"


# -------------------------------
# 1. PyPDF
# -------------------------------
def extract_pypdf(path):
    text = ""
    reader = PdfReader(path)
    for page in reader.pages:
        text += page.extract_text() or ""
    return text


# -------------------------------
# 2. pdfplumber
# -------------------------------
def extract_pdfplumber(path):
    text = ""
    with pdfplumber.open(path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() or ""
    return text


# -------------------------------
# 3. PyMuPDF (fitz)
# -------------------------------
def extract_pymupdf(path):
    text = ""
    doc = fitz.open(path)
    for page in doc:
        text += page.get_text()
    return text


# -------------------------------
# 4. UnstructuredPDF
# -------------------------------
def extract_unstructured(path):
    elements = partition_pdf(filename=path)
    text = "\n".join([el.text for el in elements if el.text])
    return text


# -------------------------------
# 5. PDFMiner
# -------------------------------
def extract_pdfminer(path):
    return extract_text(path)


# -------------------------------
# Run all
# -------------------------------
if __name__ == "__main__":

    outputs = {
        "pypdf": extract_pypdf(PDF_PATH),
        "pdfplumber": extract_pdfplumber(PDF_PATH),
        # "pymupdf": extract_pymupdf(PDF_PATH),
        # "unstructured": extract_unstructured(PDF_PATH),
        "pdfminer": extract_pdfminer(PDF_PATH),
    }

    # Print preview
    for name, text in outputs.items():
        print(f"\n{'='*20} {name.upper()} {'='*20}")
        print(text[:5000])  # first 1000 chars preview

        # Save full output
        with open(f"{name}_output.txt", "w", encoding="utf-8") as f:
            f.write(text)
```

# JSONLoader

```python
import json

data = {
    "patient_name": "Alan D'Souza",
    "age": 45,
    "disease": "Bacterial Infection",
    "medicines": [
        "Azithromycin",
        "Paracetamol"
    ],
    "advantages": [
        "Fast recovery",
        "Reduced fever"
    ],
    "disadvantages": [
        "Nausea",
        "Stomach pain"
    ]
}

# Save JSON file
with open("medical_sample.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4)

print("JSON file created ✅")
```

```python
from langchain_community.document_loaders import JSONLoader
```

```python
from langchain_community.document_loaders import JSONLoader

loader = JSONLoader(
    file_path="medical_sample.json",
    jq_schema=".",          # load entire JSON
    text_content=False      # keep structured format
)

data = loader.load()

print(data)
```
