```python
import re

def clean_pdf_text(text: str) -> str:
    # Replace bullet glyph
    text = text.replace("\uf0b7", "•")

    # Remove excessive spaces
    text = re.sub(r'[ \t]+', ' ', text)

    # Remove multiple newlines
    text = re.sub(r'\n+', '\n', text)

    return text.strip()
```

## Hybrid - BM25 & Similarity Score

```python
dense_retriever = vectorstore3.as_retriever(search_kwargs={"k": 3})
```

```python
all_docs = vectorstore3.get()["documents"]
documents = [Document(page_content=text) for text in all_docs]

bm25 = BM25Retriever.from_documents(documents)
bm25.k = 3
```

```python
def hybrid_retrieve(query, k_dense=3, k_bm25=3, max_chunks=6):
    
    # Dense
    dense_docs = dense_retriever.invoke(query)
    
    # Sparse
    bm25_docs = bm25.invoke(query)
    
    # Combine
    combined = dense_docs + bm25_docs
    
    # Deduplicate by content
    seen = set()
    unique_docs = []
    
    for doc in combined:
        if doc.page_content not in seen:
            unique_docs.append(doc)
            seen.add(doc.page_content)
    
    return unique_docs[:max_chunks]
```

```python
query = "What is the abstract of the paper?"
results = hybrid_retrieve(query)
```

```python
context = "\n\n".join([doc.page_content for doc in results])

response = client2.models.generate_content(
    model="gemini-2.5-flash",
    contents=context + "\n\nQuestion: " + query
)

print(response.text)
```

```python
query = "What is the introduction of the paper paper?"
results = hybrid_retrieve(query)

context = "\n\n".join([doc.page_content for doc in results])

response = client2.models.generate_content(
    model="gemini-2.5-flash",
    contents=context + "\n\nQuestion: " + query
)

print(response.text)
```
