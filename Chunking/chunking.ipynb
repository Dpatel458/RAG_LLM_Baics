Above DOC returns LIST - due the LANGCHAIN PyPdfLoader will give LIST as output - and FIXED-LENGTH SPLITTER wants string so there are 2 solutions for fixed length splitters:
1) Use splitter as **split_documents**
2) Looping "doc" and get list to string - and then use **split_text**

**With split_documents()**

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter(
    chunk_size=100, chunk_overlap=0, separator=" "
)
texts = text_splitter.split_documents(doc)
```

Logic Breakdown:\
**Open PDF\
 ↓\
Loop over pages\
 ↓\
Extract text per page\
 ↓\
Store text + page number\
 ↓\
Return list of page chunks**

```python
from pypdf import PdfReader

def page_based_chunking(pdf_path):
    reader = PdfReader(pdf_path)
    
    chunks = []
    
    for page_number, page in enumerate(reader.pages):
        text = page.extract_text()
        
        # Clean None cases
        if text is None:
            text = ""
        
        chunk = {
            "content": text.strip(),
            "page": page_number + 1
        }
        
        chunks.append(chunk)
    
    return chunks

pdf_path = "D:/Adarta/Document_Loaders/Documents/FINAL REPORT.pdf"
chunks = page_based_chunking(pdf_path)

print("Total pages:", len(chunks))
print("\nFirst page preview:\n")
print(chunks[5]["content"][:])
print("\nMetadata for the page:\n")
print("\nPage number:", chunks[5]["content"][:100])
```

# Semantic Chunking

Logic: For semantics - we need similarity score - which can be calculated on only vectors: so use cosine similarity and then merge vectors as per a given threshold\
use **en_core_web_lg** (large) space vocab for better similarity scores

```python
%pip install https://github.com/explosion/spacy-models/releases/download/en_core_web_lg-3.7.1/en_core_web_lg-3.7.1-py3-none-any.whl
```

```python
import spacy 
nlp = spacy.load("en_core_web_lg")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def semantic_chunk_spacy(text, threshold=0.6):
    
    doc = nlp(text)
    
    sentences = [sent.text.strip() for sent in doc.sents]
    
    chunks = []
    current_chunk = [sentences[0]]
    
    for i in range(1, len(sentences)):
        
        sent1 = nlp(sentences[i-1]).vector
        sent2 = nlp(sentences[i]).vector
        
        sim = cosine_similarity([sent1], [sent2])[0][0]
        
        if sim < threshold:
            chunks.append(" ".join(current_chunk))
            current_chunk = [sentences[i]]
        else:
            current_chunk.append(sentences[i])
    
    chunks.append(" ".join(current_chunk))
    
    return chunks
```

```python
text = """
Machine learning is widely used in industry.
Neural networks are powerful deep learning models.
Backpropagation helps train neural networks.

The Indian monsoon affects agriculture heavily.
Rainfall prediction is crucial for farmers.
Climate change impacts seasonal rainfall.
"""

chunks = semantic_chunk_spacy(text, threshold=0.7)

for i, chunk in enumerate(chunks):
    print(f"\nChunk {i+1}:\n{chunk}")
```
