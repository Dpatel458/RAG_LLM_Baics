## Quotes.toscrape.com

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

# Setup driver
service = Service("chromedriver.exe")
driver = webdriver.Chrome(service=service)

driver.get("https://quotes.toscrape.com/")

# Small wait (page is static, but good habit)
time.sleep(1)

# Find all books
quotes = driver.find_elements(By.CLASS_NAME, "col-md-8")
quotes = driver.find_elements(By.CLASS_NAME, "quote")

for quote in quotes:
    text = quote.find_element(By.CLASS_NAME, "text").text
    author = quote.find_element(By.CLASS_NAME, "author").text
    
    print(text)
    print(author)
    print("-" * 40)

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

# Setup driver
service = Service("chromedriver.exe")
driver = webdriver.Chrome(service=service)

driver.get("https://books.toscrape.com/")

# Small wait (page is static, but good habit)
time.sleep(2)

# Find all books
books = driver.find_elements(By.CLASS_NAME, "product_pod")

print("Total books found:", len(books))
print("-" * 50)

for book in books:
    
    # Title
    title = book.find_element(By.TAG_NAME, "a").get_attribute("title")
    
    # Price
    price = book.find_element(By.CLASS_NAME, "price_color").text
    
    # Availability
    availability = book.find_element(By.CLASS_NAME, "instock").text.strip()
    
    # Rating (stored inside class name)
    rating_element = book.find_element(By.CLASS_NAME, "star-rating")
    rating_classes = rating_element.get_attribute("class")
    
    # Extract rating word (One, Two, Three, Four, Five)
    rating = rating_classes.split()[-1]
    
    print(f"Title: {title}")
    print(f"Price: {price}")
    print(f"Availability: {availability}")
    print(f"Rating: {rating}")
    print("-" * 50)

driver.quit()
```

# WebBaseLoader - LANGCHAIN

## Multiple Websites

```python
from langchain_community.document_loaders import WebBaseLoader

url = "https://quotes.toscrape.com/"

loader = WebBaseLoader(url)

documents = loader.load()

print("Number of documents:", len(documents))
print("-" * 50)

print("Metadata:")
print(documents[0].metadata)

print("-" * 50)
print("Page Content Preview:")
print(documents[0].page_content[:500])  # first 500 chars
```
